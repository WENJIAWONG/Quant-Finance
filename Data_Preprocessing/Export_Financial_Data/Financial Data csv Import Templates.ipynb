```python
tickers_list = ['AAPL', 'BTC-USD', 'ETH-USD', '^GSPC']
df_all = yf.download(tickers_list,'2015-1-1')['Adj Close']
print(df_all.head())
```

```python
((df_all.pct_change()+1).cumprod()).plot(figsize=(16,9))
plt.legend()
plt.title("Returns", fontsize=16)
plt.ylabel('Cumulative Returns', fontsize=14)
plt.xlabel('Year', fontsize=14)

plt.grid(which='major', color='k', linestyle='-.', linewidth=0.5)
plt.show()
```

```python
import numpy as np
start = datetime.datetime(2012,5,31) 
end = datetime.datetime(2021,9,1)
BTC = yf.download("BTC-USD", start=start, end=end, group_by='ticker')
BTC['Log_Return'] = np.log(BTC['Adj Close'] / BTC['Adj Close'].shift(1))
print(BTC['Log_Return'])
```

```python
BTC['Log_Return'].plot(figsize=(16,9))
plt.legend()
plt.title("Log Returns", fontsize=16)
plt.ylabel('Cumulative Log Returns)', fontsize=14)
plt.xlabel('Year', fontsize=14)

plt.grid(which='major', color='k', linestyle='-.', linewidth=0.5)
plt.show()
```
